```python
def mean_normalization(data, feature):
    # mean normalization(평균 정규화) -1 ~ 1
    return (data[feature] - data[feature].mean()) / data[feature].max()

def normalization(data, feature):
    return data[feature] / data[feature].max()

def preprocessing(data):
    X = data.drop(["PassengerId", "Name", "Ticket", "Cabin", "Embarked"], axis=1) # 불필요하다고 판단된 특성 제거
    X["Sex"] = (X["Sex"] == "male").astype(int) # 남자일경우 1, 여자일경우 0으로 수정
    X["Age"] = X["Age"].fillna(X["Age"].mean()) # 나이에서 Nan값은 나이 평균값으로 채워줌
    X["Age"] = normalization(X, "Age") # normalization
    X["Pclass"] = normalization(X, "Pclass")
    X["Fare"] = X["Fare"] / (X["SibSp"] + X["Parch"] + 1)
    X["Fare"] = normalization(X, "Fare")
    X["SibSp"] = normalization(X, "SibSp")
    X["Parch"] = normalization(X, "Parch")
    
    return X
```
